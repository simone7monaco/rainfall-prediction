```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

os.chdir("./fig5")
```

```python
dfBCE = pd.read_csv('BCE.csv')
dfCAPE = pd.read_csv('CAPE.csv')

dfCAPE["epoch"] = dfCAPE["epoch"] + 42

sns.lineplot(data=dfBCE, x="epoch", y="val/loss", label='BCE')
sns.lineplot(data=dfCAPE, x="epoch", y="val/loss", label='CAPE')
plt.legend()
plt.title("VAL LOSS")
```
